- **Primeiras e últimas linhas** — `head()` e `tail()` para checar se os dados fazem sentido visualmente
- **Schema/dicionário de dados** — o que cada coluna significa no contexto do negócio

Serve para: evitar surpresas durante a análise, entender o tamanho do problema e definir o que faz sentido calcular.

### 1.1 Shape
Quantas linhas e colunas existem no dataset.

### 1.2 Tipos de Variáveis
Numéricas, categóricas, datas, booleanas — define quais análises são aplicáveis a cada coluna.

### 1.3 Primeiras e Últimas Linhas
`head()` e `tail()` para verificar se os dados fazem sentido visualmente antes de qualquer cálculo.

### 1.4 Schema / Dicionário de Dados

## **2. Qualidade dos Dados**

- **Valores ausentes (missing values)** — quantidade e percentual por coluna; padrão de ausência (aleatório ou sistemático?)
- **Duplicatas** — linhas idênticas ou quase idênticas
- **Inconsistências** — valores impossíveis (ex: idade negativa, data futura em campo de nascimento)
- **Outliers brutos** — valores absurdos que indicam erro de coleta (diferente de outlier estatístico)
- **Cardinalidade de categóricas** — colunas com milhares de categorias únicas podem ser IDs disfarçados

Serve para: não construir análises sobre dados corrompidos. Lixo entra, lixo sai.

### 2.1 Valores Ausentes (Missing Values)
Quantidade e percentual por coluna; identificar se a ausência é aleatória ou sistemática orienta a estratégia de imputação.

### 2.2 Duplicatas
Linhas idênticas ou quase idênticas que distorcem distribuições e inflam métricas de frequência.

### 2.3 Inconsistências
Valores impossíveis, como idade negativa ou datas futuras em campos de nascimento.

### 2.4 Outliers Brutos
Valores absurdos que indicam erro de coleta, distintos de outliers estatísticos legítimos.

### 2.5 Cardinalidade de Categóricas
Colunas com milhares de categorias únicas podem ser IDs disfarçados, não variáveis analíticas.

## **3. Medidas de Tendência Central**

Respondem: onde os dados estão concentrados?

| Medida | Definição | Quando usar |
|--------|-----------|-------------|
| **Média** | Soma dividida pela quantidade | Distribuições simétricas, sem outliers extremos |
| **Mediana** | Valor central quando ordenado | Distribuições assimétricas (renda, preço de imóvel) |
| **Moda** | Valor mais frequente | Variáveis categóricas ou discretas |
| **Média aparada (trimmed mean)** | Média excluindo os X% extremos | Quando há outliers mas não quer usar só a mediana |

Serve para: ter um "valor típico" de referência para cada variável numérica.

### 3.1 Média
Soma dividida pela quantidade. Sensível a outliers; indicada para distribuições simétricas.

### 3.2 Mediana
Valor central quando os dados estão ordenados. Robusta a outliers; preferível em distribuições assimétricas.

### 3.3 Moda
Valor mais frequente. Única medida de tendência central aplicável a variáveis categóricas.

### 3.4 Média Aparada (Trimmed Mean)
Média calculada excluindo os X% extremos — equilíbrio entre a média e a mediana quando há outliers.

## **4. Medidas de Dispersão**

Respondem: os dados estão espalhados ou concentrados?

| Medida | Definição | Quando usar |
|--------|-----------|-------------|
| **Variância** | Média dos quadrados dos desvios | Base teórica, mas difícil de interpretar |
| **Desvio padrão** | Raiz da variância, mesma unidade dos dados | Principal medida de dispersão |
| **Amplitude** | Máximo − Mínimo | Rápido, mas sensível a outliers |
| **IQR (Intervalo Interquartil)** | Q3 − Q1 (50% centrais) | Robusto a outliers, base do boxplot |
| **CV (Coeficiente de Variação)** | Desvio padrão / Média | Comparar dispersão entre variáveis em escalas diferentes |

Serve para: entender se a média é representativa ou se os dados são muito heterogêneos.

### 4.1 Variância
Média dos quadrados dos desvios em relação à média. Base teórica de outras métricas de dispersão.

### 4.2 Desvio Padrão
Raiz quadrada da variância, expressa na mesma unidade dos dados. Medida de dispersão mais interpretável.

### 4.3 Amplitude
Máximo − Mínimo. Simples e rápida de calcular, porém muito sensível a outliers.

### 4.4 IQR (Intervalo Interquartil)
Q3 − Q1, representando os 50% centrais dos dados. Robusto a outliers e base do boxplot.

### 4.5 CV (Coeficiente de Variação)
Desvio padrão dividido pela média. Permite comparar dispersão entre variáveis com escalas diferentes.

## **5. Distribuição e Forma**

Respondem: qual o formato da distribuição dos dados?

- **Histograma** — visualiza a forma da distribuição
- **Boxplot** — mostra mediana, IQR e outliers de forma compacta
- **KDE (Kernel Density Estimation)** — versão suavizada do histograma
- **Assimetria (Skewness)** — distribuição puxada para a direita (positiva) ou esquerda (negativa)
- **Curtose (Kurtosis)** — cauda mais pesada ou mais leve que a normal (detecta risco de outliers extremos)
- **Teste de normalidade** — Shapiro-Wilk, Kolmogorov-Smirnov (muitos modelos assumem normalidade)

Serve para: escolher as transformações corretas (log, sqrt), identificar o tipo certo de teste estatístico e entender riscos de outliers.

### 5.1 Histograma
Divide os valores em intervalos e conta a frequência de cada faixa.

### 5.2 Boxplot
Representa visualmente mediana, IQR, mínimo, máximo e outliers de forma compacta.

### 5.3 KDE (Kernel Density Estimation)
Estimativa suavizada da densidade de probabilidade — versão contínua do histograma, útil para comparar distribuições sobrepostas.

### 5.4 Assimetria (Skewness)
Mede o grau e a direção da assimetria da distribuição em torno da média.

### 5.5 Curtose (Kurtosis)
Mede o peso das caudas — distribuições leptocúrticas têm caudas pesadas e maior risco de outliers extremos.

### 5.6 Teste de Normalidade
Shapiro-Wilk e Kolmogorov-Smirnov verificam formalmente se a distribuição é normal.

## **6. Análise de Outliers**

- **Método IQR** — valores abaixo de Q1 − 1.5×IQR ou acima de Q3 + 1.5×IQR
- **Z-score** — valores com |z| > 3 (mais de 3 desvios padrões da média)
- **Isolation Forest / LOF** — métodos de ML para outliers multivariados
- **Visualização** — boxplot, scatter plot com cores, violin plot

Serve para: decidir se o outlier é erro, evento raro legítimo ou informação valiosa. A decisão impacta direto na modelagem.

### 6.1 Método IQR
Considera outlier qualquer valor abaixo de Q1 − 1.5×IQR ou acima de Q3 + 1.5×IQR.

### 6.2 Z-Score
Considera outlier qualquer valor com |z| > 3, ou seja, além de 3 desvios padrões da média.

### 6.3 Isolation Forest / LOF
Algoritmos de ML que detectam outliers multivariados sem depender de uma única dimensão.

### 6.4 Visualização
Boxplot, scatter plot com destaque por cor e violin plot para identificar outliers visualmente.

- **Gráfico de barras** — visualização da distribuição
- **Detecção de categorias raras** — categorias com < 1% de representatividade podem ser agrupadas
- **Análise de cardinalidade** — variáveis com alta cardinalidade exigem encoding cuidadoso

Serve para: entender o desbalanceamento entre categorias, que afeta modelos de classificação e análises comparativas.

### 7.1 Frequência Absoluta e Relativa
Contagem e percentual de cada categoria na variável.

### 7.2 Gráfico de Barras
Visualização da distribuição entre categorias, ordenável por frequência para facilitar a leitura.

### 7.3 Detecção de Categorias Raras
Categorias com representatividade abaixo de 1% podem ser agrupadas em "outros" para reduzir ruído.

### 7.4 Análise de Cardinalidade
Número de categorias únicas — alta cardinalidade exige encoding cuidadoso antes da modelagem.

## **8. Análise Bivariada**

Responde: como duas variáveis se relacionam entre si?

**Numérica × Numérica**
- **Scatter plot** — visualiza a relação e padrões
- **Correlação de Pearson** — relação linear (sensível a outliers)
- **Correlação de Spearman** — relação monotônica, robusta a outliers
- **Correlação de Kendall** — melhor para amostras pequenas

**Numérica × Categórica**
- **Boxplot agrupado** — compara distribuições por grupo
- **Violin plot** — mostra a densidade por grupo
- **Teste t / ANOVA** — verifica se as médias diferem estatisticamente entre grupos

**Categórica × Categórica**
- **Tabela de contingência (crosstab)** — frequência cruzada
- **Heatmap de frequência** — visualização da tabela
- **Teste qui-quadrado (χ²)** — verifica independência entre categóricas

Serve para: encontrar relações que guiarão a engenharia de features e a escolha do modelo.

### 8.1 Numérica × Numérica
Scatter plot, correlação de Pearson (linear), Spearman (monotônica) e Kendall (amostras pequenas).

### 8.2 Numérica × Categórica
Boxplot agrupado, violin plot e testes t / ANOVA para verificar se as médias diferem entre grupos.

### 8.3 Categórica × Categórica
Tabela de contingência (crosstab), heatmap de frequência e teste qui-quadrado (χ²) para verificar independência.

## **9. Análise Multivariada**

Responde: como várias variáveis interagem ao mesmo tempo?

- **Heatmap de correlação** — matriz de correlações de todas as numéricas de uma vez
- **Pair plot (Scatter Matrix)** — scatter de todos os pares de variáveis, com distribuição na diagonal
- **PCA (Análise de Componentes Principais)** — reduz dimensionalidade mantendo a máxima variância; útil para visualizar clusters em 2D
- **t-SNE / UMAP** — redução não-linear para visualização de grupos complexos
- **Cluster analysis exploratório (K-Means, DBSCAN)** — identifica grupos naturais antes de qualquer modelagem supervisionada
- **Análise de multicolinearidade (VIF)** — detecta variáveis redundantes que prejudicam regressões

Serve para: entender a estrutura global dos dados, detectar redundâncias e grupos ocultos.

### 9.1 Heatmap de Correlação
Matriz visual com o coeficiente de correlação entre todas as variáveis numéricas de uma vez.

### 9.2 Pair Plot (Scatter Matrix)
Grade de scatter plots para todos os pares de variáveis, com a distribuição individual na diagonal.

### 9.3 PCA (Análise de Componentes Principais)
Reduz dimensionalidade preservando a máxima variância — útil para visualizar clusters em 2D.

### 9.4 t-SNE / UMAP
Técnicas de redução não-linear para visualizar agrupamentos em datasets de alta dimensionalidade.

### 9.5 Cluster Analysis Exploratório (K-Means, DBSCAN)
Identifica agrupamentos naturais nos dados antes de qualquer modelagem supervisionada.

### 9.6 Análise de Multicolinearidade (VIF)
Variance Inflation Factor detecta variáveis redundantes que prejudicam modelos de regressão.

## **10. Análise Temporal (se houver datas)**

- **Série temporal** — evolução de métricas ao longo do tempo
- **Sazonalidade** — padrões que se repetem (diário, semanal, anual)
- **Tendência** — crescimento ou queda estrutural
- **Decomposição de série** — separa tendência + sazonalidade + resíduo
- **Lag features** — correlação de uma variável com ela mesma em períodos anteriores (autocorrelação)

Serve para: evitar data leakage em modelos temporais e entender padrões de comportamento ao longo do tempo.

### 10.1 Série Temporal
Visualização da evolução de uma métrica ao longo do tempo.

### 10.2 Sazonalidade
Padrões cíclicos que se repetem em intervalos regulares — diário, semanal ou anual.

### 10.3 Tendência
Direção geral de crescimento ou queda estrutural ao longo do período analisado.

### 10.4 Decomposição de Série
Separa a série em três componentes: tendência, sazonalidade e resíduo.

### 10.5 Lag Features
Correlação de uma variável com versões defasadas de si mesma — base para detecção de autocorrelação.

## **12. Resumo e Hipóteses**

A etapa final da EDA não é técnica, é interpretativa.

- **Listar achados principais** — o que chamou atenção?
- **Levantar hipóteses** — "parece que X influencia Y por causa de Z"
- **Definir o que precisa ser tratado** — missings, outliers, transformações, encodings
- **Documentar decisões** — por que você escolheu tratar (ou não) cada problema encontrado

Serve para: transformar números e gráficos em entendimento de negócio. É o que diferencia uma EDA rasa de uma análise que realmente orienta decisões.

### 12.1 Listar Achados Principais
Registrar o que chamou atenção — variáveis com comportamento inesperado, correlações fortes, desequilíbrios.

### 12.2 Levantar Hipóteses
Formalizar intuições geradas pela análise — "parece que X influencia Y por causa de Z".

### 12.3 Definir o que Precisa ser Tratado
Consolidar o backlog de pré-processamento — missings, outliers, transformações, encodings.

### 12.4 Documentar Decisões
Registrar por que cada problema foi ou não tratado, criando rastreabilidade para o modelo.
